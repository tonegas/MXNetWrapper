```python
from keras import models
from keras import layers
from keras import regularizers
import numpy as np
```

## Network Strucure

```python
data_in = layers.Input(shape=(2,),name='in')
lay1     = layers.Dense(50, activation='sigmoid')(data_in)
lay2     = layers.Dense(20, activation='relu')(lay1)
data_out  = layers.Dense(2,name='out',activation='tanh')(lay2)
net  = models.Model(inputs=data_in, outputs=data_out)
```

```python
net.compile(loss='mse',
             optimizer='Adam')
```

```python
# first train model
# define train set
x_train = np.empty((10,2))
for i in range(0,10):
    x_train[i] = np.array((i,i*i)).flatten()

# define train set
y_train = np.empty((10,2))
for i in range(0,10):
    y_train[i] = np.array((1,0)).flatten()
    
hist0 = net.fit(x_train, y_train,
                
         epochs=500,
         batch_size=128,
         verbose=2);
```

```python
net.predict(np.array([(3,1),(2,3)]))
```

```python
def export_model(net,name,epoch=0):
    from keras import backend as K
    backend = K.backend()
    if backend == 'tensorflow':
        # export model to hdf5 file
        net.save(name+".h5")
        jsonNet= net.to_json()
        with open(name+".json","w") as json_file:
            json_file.write(net)
    elif backend == 'mxnet':
        from keras.models import save_mxnet_model
        data_names, data_shapes = save_mxnet_model(model=net, prefix=name, epoch=epoch)
```

```python
export_model(net,'test')
```

```python
def import_model(name, data_in_name, data_in_shape, epoch=0):
    from keras import backend as K
    backend = K.backend()
    if backend == 'tensorflow':
        pass
    elif backend == 'mxnet':
        import mxnet as mx
        sym, arg_params, aux_params = mx.model.load_checkpoint(prefix=name, epoch=epoch)
        net = mx.module.Module(symbol = sym,
                               data_names = data_in_name,
                               context = mx.cpu(),
                               label_names = None)
        net.bind(for_training=False,
                 data_shapes=data_in_shape)
        net.set_params(arg_params, aux_params, allow_missing=True)
        return net

    
# For version of mxnet lower the 1.3.0
def export_fix(name, epoch=0):
    import mxnet as mx
    sym, arg_params, aux_params = mx.model.load_checkpoint(prefix=name, epoch=epoch)
    # export fixed net for MXNet C++ loading
    fix_arg_params = {}
    for i, key in enumerate(arg_params):
        fix_arg_params['arg:'+key]=arg_params[key]

    mx.nd.save("testFix-0000.params",fix_arg_params)
```

```python
loadednet=import_model("test",["/in1"],[("/in1",(2,))])
loadednet.predict(np.array([(2,4),(2,3)]))
```
